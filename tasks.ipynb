```python
import math
import numpy as np  # Импортируем библиотеку NumPy
import matplotlib.pyplot as plt 

a = -1
b = 1
def func(x):
    # return 1/(1+ 25 * x**2)
    return math.exp(x)

def polyLagrange(x, n):
    nodes = [a]
    current = a
    for k in range(n):
        current += 2 / n
        nodes.append(current)
    nodes.sort()

    values = list(map(math.exp, nodes))
    
    summa = 0
    
    for x_val in x:  # Изменил имя переменной для ясности
        for i in range(n + 1):
            # Вычисление произведения
            product_up = 1
            for j in range(n + 1):
                if i != j:
                    product_up *= (x_val - nodes[j])
            product_down = 1
            for j in range(n + 1):
                if i != j:
                    product_down *= (nodes[i] - nodes[j])
            summa += (values[i] * (product_up / product_down))
    
    return summa

n = 3




x = np.linspace(-1, 1, n + 1)  # n+1 точек от -1 до 1
y = [polyLagrange([x_val], n) for x_val in x]  # Применение функции к каждой точке x

# Интерполяция с большим числом узлов
x_2 = np.linspace(-1, 1, 2 * n + 2)  # 2n+2 точки от -1 до 1
y_2 = [polyLagrange([x_val], n) for x_val in x_2]  # Использовать x_2 для вычисления y_2

# Интерполяция с большим числом узлов
x_3 = np.linspace(-1, 1, 10 * n+1)  # 2n+2 точки от -1 до 1
y_3 = [polyLagrange([x_val], n) for x_val in x_3]  # Использовать x_2 для вычисления y_2

# Создаем график
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Интерполяция в узлах', color='blue')  
plt.plot(x_2, y_2, label='Интерполяция в узлах и средних точках', color='green')  
plt.plot(x_3, y_3, label='Приближаемая функция', color='red')  

plt.title('График интерполяции Лагранжа')  
plt.xlabel('x')  
plt.ylabel('f(x)')  
plt.axhline(0, color='black', linewidth=0.5, ls='--')  
plt.axvline(0, color='black', linewidth=0.5, ls='--')  
plt.grid(True)  
plt.legend()  
# plt.ylim(-10, 10)  # Ограничение по оси y
plt.show()  # Показываем график 

```
